## IMPORT LIBRARY

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
#202331085_Miqdad Hilya Hasan
```

## MEMBACA GAMBAR

```python
img = cv2.imread('harimau_siberia.jpg')
#202331085_Miqdad Hilya Hasan
```

```python
img.shape
#202331085_Miqdad Hilya Hasan
```

```python
[baris, kolom] = img.shape[:2]
#202331085_Miqdad Hilya Hasan
```

## CONVERT BGR TO RGB

```python
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#202331085_Miqdad Hilya Hasan
```

## MENAMPILKAN HISTOGRAM

```python
fig, axs = plt.subplots(2, 2, figsize=(15,5))

#Cara 1
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])

#Cara 2
hist = cv2.calcHist([rgb],[0], None,[256],[0,256])
axs[1,0].imshow(rgb)
axs[1,1].plot(hist)

#Menampilkan
plt.show
#202331085_Miqdad Hilya Hasan
```

## OPERASI PIKSEL UNTUK MENINGKATKAN KECERAHAN

```python
beta = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] + beta
        citra_cerah[x,y] = gyx

citra_cerah = citra_cerah.astype(np.uint8)

fix, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256, [0,256])
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256])

plt.show()
#202331085_Miqdad Hilya Hasan
```

## MENGUBAH KONTRAS

```python
alpha = 1.2
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] * alpha
        citra_kontras[x,y] = gyx

citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])

plt.show()
#202331085_Miqdad Hilya Hasan
```

## OPERASI PIKSEL GABUNGAN ANTARA MENINGKATKAN KECERAHAN DAN MEREGANGKAN KONTRAS

```python
beta  = 10
alpha = 1.2
citra_hasil = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = alpha * rgb[x,y] + beta
        citra_hasil[x,y] = gyx

citra_hasil = citra_hasil.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_hasil)
axs[1,1].hist(citra_hasil.ravel(),256,[0,256])

plt.show()
#202331085_Miqdad Hilya Hasan
```

## MEMBALIK CITRA

```python
beta = 10
alpha = 1.2
citra_negatif = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        fyx = 255 - rgb[x,y]
        citra_negatif[x,y] = fyx

citra_negatif = citra_negatif.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_negatif)
axs[1,1].hist(citra_negatif.ravel(),256,[0,256])

plt.show()
#202331085_Miqdad Hilya Hasan
```

```python

fig, axs = plt.subplots(5,2, figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(),256,[0,256])
axs[1,0].imshow(citra_negatif)
axs[1,1].hist(citra_negatif.ravel(),256,[0,256])
axs[2,0].imshow(citra_cerah)
axs[2,1].hist(citra_cerah.ravel(),256,[0,256])
axs[3,0].imshow(citra_kontras)
axs[3,1].hist(citra_kontras.ravel(),256,[0,256])
axs[4,0].imshow(citra_hasil)
axs[4,1].hist(citra_hasil.ravel(),256,[0,256])

plt.show()
#202331085_Miqdad Hilya Hasan
```
